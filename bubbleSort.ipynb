### Bubble Sort
It is a sorting algorithm where starting from the beginning each item is compared with the following item, if it is larger switch the places. But each time array that will be sorted is going to be n-1 length array, since the last one is already been compared with the rest. It takes o(n^2) time.

```python
from array import array
from timeit import timeit
import matplotlib.pyplot as plt
import random
```

```python
def gen_arr(size):
    return array("i", [i for i in range(size)])
```

```python
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(len(arr) - 1 - i):
            if(arr[j] > arr[j + 1]):
                temp = arr[j]
                arr[j] = arr[j + 1]
                arr[ + 1] = temp
```

```python
def test_bubble():
    sizes = [n for n in range(0,300,10)]
    arrays = list(map(gen_arr, sizes[1:]))
    search_operation = lambda: bubble_sort(arr)
    time_taken = []
    timeDict = {}
    for size, arr in zip(sizes, arrays):
        time_taken = timeit(search_operation, number = 100)
        timeDict[size] = time_taken
    return timeDict
```

```python
timeDict = test_bubble()
```

```python
plt.plot(list(timeDict.keys()), list(timeDict.values()), marker = 'o')
plt.title('Time taken for Linear Search')
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.show()
```
