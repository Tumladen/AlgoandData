### Binary Search Algorithm
It is a nice algorithm, given the array is ordered. What it does is splits the array into two parts, then checks if the value is smaller or larger than what is being searched for, according to that divides the subarray into two parts where the value is in, then those subarrays, so on, until only two item is left to be compared, unless the values hasn't been found yet. It takes o(log n) time.

```python
from array import array
from timeit import timeit
import matplotlib.pyplot as plt
```

```python
def gen_arr(size):
    return array("i", [i for i in range(size)])
```

```python
def binary_search(arr, val):
    lo = 0
    hi = len(arr)
    while lo < hi:
        mid = lo + (hi - lo) // 2
        mid_val = arr[mid]
        if mid_val == val:
            return True
        elif mid_val > val:
            hi = mid
        else:
            lo = mid + 1
    return False
```

```python
def test_binary():
    sizes = [n for n in range(0,10000, 10)]
    arrays = list(map(gen_arr, sizes[1:]))
    search_operation = lambda: binary_search(arr, arr[-1])
    timeDict = {}
    for size, arr in zip(sizes, arrays):
        time_taken = timeit(search_operation, number = 100)
        timeDict[size] = time_taken
    return timeDict
```

```python
timeDict = test_binary()
```

```python
plt.plot(list(timeDict.keys()), list(timeDict.values()), marker = 'o')
plt.title('Time taken for Binary Search')
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.show()
```
