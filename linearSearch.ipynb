### Linear Search
The most basic type of search. It takes o(n) time. It starts from the beginning runs through the entire array.

```python
from array import array
from timeit import timeit
import matplotlib.pyplot as plt
```

```python
def gen_arr(size):
    return array("i", [i for i in range(size)])
```

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return True
    return False
```

```python
def test_linear():
    sizes = [n for n in range(0,10000, 10)]
    arrays = list(map(gen_arr, sizes[1:]))
    search_operation = lambda: linear_search(arr, arr[-1])
    time_taken = []
    timeDict = {}
    for size, arr in zip(sizes, arrays):
        time_taken = timeit(search_operation, number = 100)
        timeDict[size] = time_taken
    return timeDict
```

```python
timeDict = test_linear()
```

```python
plt.plot(list(timeDict.keys()), list(timeDict.values()), marker = 'o')
plt.title('Time taken for Linear Search')
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.show()
```
